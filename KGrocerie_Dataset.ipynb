```python
# detect the outlier

import numpy as np
import pandas as pd
from scipy.stats import median_abs_deviation

# Assuming your DataFrame is named 'df' and has already been loaded

# Calculate the median absolute deviation (MAD) for numeric columns
mad = median_abs_deviation(df.select_dtypes(include=np.number), axis=0)

# Define a threshold for outlier detection (e.g., 3 times the MAD)
threshold = 3 * mad

# Identify outliers
outliers = df[abs(df.select_dtypes(include=np.number) - df.select_dtypes(include=np.number).median()) > threshold]


# Print or further analyze the outliers
print("Outliers:")
outliers
```

```python
df = pd.read_csv('cleaned_data.csv')

# Convert dataset to DataFrame with one-hot encoding
df = pd.get_dummies(df, columns=['itemdescription'], prefix='item')
df.head(25)
```

```python
#Encode the Transactions
from mlxtend.preprocessing import TransactionEncoder  # Import TransactionEncoder

te = TransactionEncoder()  # Correct the class name
transactions_matrix = te.fit_transform(transactions)
transactions_df = pd.DataFrame(transactions_matrix, columns=te.columns_)

# Assuming 'transactions_df' is the encoded DataFrame, change 'encoded_df' to 'transactions_df'
print(transactions_df.head())
```

```python
print(len(transactions_df))
```

```python
transactions_df = transactions_df.astype(int)
transactions_df.head(10)
```

Apriori

```python
#run apriori to get frequent itemsets
import time # Import the time module

start_time = time.time()
from mlxtend.frequent_patterns import apriori # Import apriori function
freqitems_apriori = apriori(transactions_df, min_support=0.001, use_colnames=True)
print("Apriori execution time: {:.2f} seconds".format(time.time() - start_time))
```

```python
freqitems_apriori.sort_values(by= "support",ascending=False)
print(freqitems_apriori.sort_values(by= "support",ascending=False))
```

```python
#Generate association rules from frequent itemsets
from mlxtend.frequent_patterns import association_rules # Import association_rules
rules = association_rules(freqitems_apriori, metric="confidence", min_threshold=0.1,num_itemsets=3)
print(rules.head(25))
```

```python
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head(25))
```

FP-Growth¶


```python
#run fp-growth to get frequent itemsets
import time # Import the time module

start_time = time.time()
from mlxtend.frequent_patterns import fpgrowth
freqitems_fpgrowth = fpgrowth(transactions_df, min_support=0.001, use_colnames=True)
print("FP-Growth execution time: {:.2f} seconds".format(time.time() - start_time))
print(freqitems_fpgrowth.head(25))
```

```python
freqitems_fpgrowth.sort_values(by= "support",ascending=False)
```

```python
rules_fpgrowth = association_rules(freqitems_fpgrowth, metric='confidence', min_threshold=0.1,num_itemsets=3) # Changed 'feqitems_fp' to 'freqitems_fpgrowth'
rules_fpgrowth.head(10)
```

```python


from matplotlib import pyplot as plt
rules_fpgrowth['antecedent support'].plot(kind='line', figsize=(8, 4), title='antecedent support')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
rules_fpgrowth=rules_fpgrowth.sort_values(['confidence', 'lift'], ascending =[False, False])
rules_fpgrowth
```

```python
# Create two subplots for Apriori and FP-Growth
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
# Plot for Apriori
rules_fpgrowth['antecedent support'].plot(kind='line', ax=axes[0], title='antecedent support')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')
axes[0].spines[['top', 'right']].set_visible(False)
# Plot for FP-Growth
rules_fpgrowth['consequent support'].plot(kind='line', ax=axes[1], title='consequent support')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')
axes[1].spines[['top', 'right']].set_visible(False)
#
plt.tight_layout()
plt.show()
```

```python
# Create a figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
# Plot for Apriori
rules_fpgrowth['support'].plot(kind='line', ax=axes[0], title='support')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')
axes[0].spines[['top', 'right']].set
# Plot for FP-Growth
rules_fpgrowth['lift'].plot(kind='line', ax=axes[1], title='lift')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')
axes[1].spines[['top', 'right']].set_visible(False)
#
plt.tight_layout()
plt.show()
```

```python
#run apriori to get frequent itemsets
import time # Import the time module

start_time = time.time()
from mlxtend.frequent_patterns import apriori # Import apriori function
freqitems_apriori = apriori(transactions_df, min_support=0.001, use_colnames=True)
apriori_time = time.time() - start_time # Store Apriori execution time
print("Apriori execution time: {:.2f} seconds".format(apriori_time))


#run fp-growth to get frequent itemsets
start_time = time.time()
from mlxtend.frequent_patterns import fpgrowth
freqitems_fpgrowth = fpgrowth(transactions_df, min_support=0.001, use_colnames=True)
fpgrowth_time = time.time() - start_time # Store FP-Growth execution time
print("FP-Growth execution time: {:.2f} seconds".format(fpgrowth_time))

# Plotting execution time comparison
plt.figure(figsize=(8, 5))
plt.bar(['Apriori', 'FP-Growth'], [apriori_time, fpgrowth_time], color=['blue', 'green'])
plt.ylabel('Execution Time (seconds)')
plt.title('Apriori vs FP-Growth Execution Time Comparison')
plt.show()
```
